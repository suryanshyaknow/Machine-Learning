```python
print(data.keys())

# Readying the dataframe
df = pd.DataFrame(data["data"], columns=data.feature_names)
df["price"] = data.target
df.head()
```

```python
## Check for the outliers

plt.figure(figsize=(15,10))
sns.boxplot(data=df)
plt.show()
```

```python
## features
X = df.drop(columns=["price"])

## label
y = df["price"]
```

```python
## train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
## Standardization of features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled
```

### # Model Training:

```python
from sklearn.linear_model import LinearRegression

linear_mod = LinearRegression()
linear_mod
```

```python
## fitting or training

linear_mod.fit(X_train_scaled, y_train)
```

```python
## Coefficients and Intercepts

print(linear_mod.coef_)
linear_mod.intercept_
```

```python
def pipeline(x_features, training=True):
    
    # Standardization of features
    if training:
        scaler.fit(x_test)
    return scaler.transform(x_features)
```

```python
## Predictions for the test set

y_pred = linear_mod.predict(pipeline(x_features=X_test, training=False))
y_pred
```

## # Assumptions to be made in Linear Regression:

To check whether our model is good enough or not.

- #### Predictions and True Values should follow a linear fashion.
- #### Residuals' distribution should be `Gaussian`, emphasis on mean=0.
- #### Distribution: Residuals vs Prediction, should be uniform i.e. `Homoscedasticity`.<br><br>=> `Error terms must show constant variance`.

```python
# 1 

sns.scatterplot(y_test, y_pred)
plt.xlabel("y_test")
plt.ylabel("predictions")
plt.show()
```

```python
# 2

residuals = y_test-y_pred
sns.displot(residuals, kind='kde')
```

=> Skewness in the right must've been due to the outliers.

```python
# 3 
# Error terms must show constant variance.
# Scatterplot: predictions vs residuals => UNIFORM DISTRIBUTION

sns.scatterplot(x=y_pred, y=residuals)
plt.xlabel("predictions")
plt.ylabel("residuals")
```

### # Performance Metrics:

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error


print("MSE: ", mean_squared_error(y_test, y_pred))
print("MAE: ", mean_absolute_error(y_test, y_pred))
print("RMSE: ", np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
## R-squared and Adjusted r-squared
from sklearn.metrics import r2_score

print("Score on training set: ", linear_mod.score(X_train_scaled, y_train))
print("R-squared: ", linear_mod.score(pipeline(X_test, training=False), y_test))
print("R-squared: ", r2_score(y_test, y_pred))
```

```python
## Adjusted R-squared

def adjusted_r2(r2, df:pd.DataFrame=df):
    n = df.shape[0]
    p = df.shape[1]-1
    return (1 - ((1-r2)*(n-1)/(n-p-1)))

print("adjusted R-squared: ", adjusted_r2(r2=r2_score(y_test, y_pred)))
```
