```python
## Remove irrelevant words

from nltk.corpus import stopwords

stopwords.words('english')
```

### Punctuations:

```python
import string

punches = string.punctuation
punches += '"'
punches += '``'
punches += "''"

punches
```

### # Remove all the `stopwords` and `punctuations` from the `text`:

```python
text
```

### Cleansing:

```python
text_cleansed = []
for word in nltk.word_tokenize(text):
    if word not in punches:  # filtered punctuations
        if word not in stopwords.words('english'):
            text_cleansed.append(word)
```

```python
text_cleansed
```

### # PartsOfSpeech for `text_cleansed`:

```python
nltk.pos_tag(text_cleansed)
```

### # Stemmers:

```python
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
```

**`Stemmer`** is to stem (or put some restrictions)  on the same kinda words like "ruin" and "ruining" and to consider them one rather than different that is by default.

```python
porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer('english')
```

```python
porter.stem('ruining')
```

```python
porter.stem("ruin")
```

```python
porter.stem("shameless")
```

```python
print(porter.stem("classification"))
print(lancaster.stem("classification"))
print(snowball.stem("classification"))
```

```python
print(porter.stem("regression"))
print(lancaster.stem("regression"))
print(snowball.stem("regression"))
```

**=>** Chopping the -ing, -s, -es kinda words and returning the `base` words.

### # Lemmatizer:

Unlike Stemmers, `Lemmatizers` will actually transform the word at hand into its base form.

```python
from nltk.stem import WordNetLemmatizer

lemma = WordNetLemmatizer()
lemma
```

```python
print(lemma.lemmatize("go"))
print(lemma.lemmatize("went"))
print(lemma.lemmatize("gone"))
```

```python
## by assigning the POS

print(lemma.lemmatize("go", pos='v'))
print(lemma.lemmatize("went", pos='v'))
print(lemma.lemmatize("gone", pos='v'))
```

```python
lemma.lemmatize('figurative', pos='v')
```
